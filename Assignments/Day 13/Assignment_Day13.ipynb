```python
#prorgam to print all the alphabets that never appear double in sequence(back to back) in sowpods.txt

file = open(r'C:\Users\lenovo\Desktop\PYTHON(INTERNSHIP)\Assignments\Day 13\sowpods.txt')
words = file.readlines()
upper ='ABCDEFGHIJKLMNOPQRSTUVWXYZ'

print("The alphabets that never appeared double:")
for alpha in upper:
    flag = False
    for w in words:
        if alpha*2 in w:
            flag = True
            break
    if(flag == False):
        print(alpha)
file.close()
```

```python
#prorgam to print all the alphabets that never appear double in sequence(back to back) in sowpods.txt

import mod1
import string

print("The alphabets that never appeared double:")
for alpha in string.ascii_uppercase:
    flag = False
    for w in mod1.words:
        if alpha*2 in w:
            flag = True
            break
    if(flag == False):
        print(alpha)
```
