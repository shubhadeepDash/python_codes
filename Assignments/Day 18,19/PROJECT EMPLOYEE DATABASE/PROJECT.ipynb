```python
#menu driven progranm

def add_employee(emp_id, emp_name, emp_DOJ, emp_desig, emp_sal):
    file = open("m_add_employee.txt","a")
    L = ["Employee_ID:",emp_id,"\t\tEmployee_Name:",emp_name,"\t\tEmployee_DOJ:",emp_DOJ,"\t\tEmployee_Designation:",emp_desig,"\t\tEmployee_Salary:",emp_sal,"\n"]
    file.writelines(L)
    file.close()
        
def add_HR(emp_id, hr_name, hr_dept, hr_role):
    file = open("m_add_newhr.txt","a")
    L = ["Employee_ID:",emp_id,"\t\tEmployee_Name:",hr_name, "\t\tEmployee_Department:",hr_dept,"\t\tEmployee_Role:",hr_role,"\n"]
    file.writelines(L)
    file.close()
     
def search_emp(emp_id):
    file = open("m_add_employee.txt","r")
    w = file.readlines()
    flag = 0
    for i in w:
        if emp_id in i:
            flag = 1
            i = i.replace("\t\t","\n")
            print(i)
    if(flag == 0):
        print("No such employee found.")
    file.close()
    
def search_hr(emp_id):
    file = open("m_add_newhr.txt","r")
    w = file.readlines()
    flag = 0
    for i in w:
        if emp_id in i:
            flag = 1
            i = i.replace('\t\t','\n')
            print(i)
    if(flag == 0):
        print("No such employee found.")
    file.close()

def upd_emp(emp_id):
    file = open("m_add_employee.txt","r")
    w = file.readlines()
    n_emp_desig = input("Enter the new designation: ")
    n_emp_salary = input("Enter the new salary: ")
    lst = [ ]
    for l in w:
        if emp_id in l:
            lst = list(l.split("\t\t"))
            print("The old details of the employee are:")
            print(lst[3])
            print(lst[4])
            lst[3] = "Employee_Designation:" + n_emp_desig
            lst[4] = "Employee_Salary:" + n_emp_salary + "\n"
        
    s = ''
    for j in lst:
        s = s + j + "\t\t"
    file.close()
    
    file = open("m_add_employee.txt","w")
    lst2 = []
    s = s.strip("\t\t")
    for n in w:
        if emp_id in n:
            lst2.append(s)
        else:
            lst2.append(n)
    file.writelines(lst2)
    file.close()
    
#driver's code
print("-----------------MENU DRIVEN EMPLOYEE DATABASE----------------")
choice = 0
while (choice != 6):
    print("\n")
    print("ENTER 1 - TO ADD EMPLOYEE")
    print("ENTER 2 - TO ADD HR MEMBER")
    print("ENTER 3 - TO SEARCH AN EMPLOYEE")
    print("ENTER 4 - TO SEARCH HR MEMBER")
    print("ENTER 5 - TO UPDATE INFORMATION OF AN EMPLOYEE")
    print("ENTER 6 - TO EXIT THE PROGRAM")
    choice = int(input())
    if choice == 1:
            emp_id = input("Enter the ID of employee:")
            emp_name = input("Enter the name of the employee:")
            emp_DOJ = input("Enter the Date of Joining of the employee:")
            emp_desig = input("Enter the designation of the employee:")
            emp_sal = "0" # salary was to be taken zero by default
            add_employee(emp_id, emp_name, emp_DOJ, emp_desig, emp_sal)
    elif choice == 2:
            emp_id = input("Enter the ID of employee:")
            hr_name = input("Enter the name of the employee:")
            hr_dept = input("Enter the Department of the employee:")
            hr_role = input("Enter the role of the employee:")
            add_HR(emp_id, hr_name, hr_dept, hr_role)
    elif choice == 3:
            emp_id = input("Enter the employee ID: ")
            search_emp(emp_id)
    elif choice == 4:
            emp_id = input("Enter the employee ID: ")
            search_hr(emp_id)
    elif choice == 5:
            emp_id = input("Enter the id of the employee to be updated:")
            upd_emp(emp_id)
```
